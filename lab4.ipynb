```python
# 20

import seaborn as sns
iris = sns.load_dataset('iris')

# minimum value of sepal_width for "virginica" variety in the iris dataset
min_sepal_width_virginica = iris[iris['species'] == 'virginica']['sepal_width'].min()
print(min_sepal_width_virginica)

# manually double check
# load all the data for me to see of sepal_width for "virginica" variety in the iris dataset
print(iris[iris['species'] == 'virginica'])
```

```python
# 21
# Construct a dataframe where petal_length is greater than the mean petal_length of whole iris dataset.

# Calculate the mean petal_length of the whole iris dataset
mean_petal_length = iris['petal_length'].mean()

# Construct a dataframe where petal_length is greater than the mean petal_length
iris_petal_length_gt_mean = iris[iris['petal_length'] > mean_petal_length]

# Display the head of the new dataframe
display(iris_petal_length_gt_mean.head())

# Count and print the number of observations in the new dataframe
num_observations = len(iris_petal_length_gt_mean)
print(f"Number of observations where petal_length is greater than the mean: {num_observations}")

# How many observations are there in this dataframe?
print(f"Number of observations in the new dataframe: {num_observations}")
```

```python
# 22

# count how many versicolor there are
versicolor_count = iris[iris['species'] == 'versicolor'].shape[0]
print(f"Number of versicolor observations: {versicolor_count}")

# find the mean value of sepal_length in this dataframe rounded to three decimal places
mean_sepal_length = round(iris[iris['species'] == 'versicolor']['sepal_length'].mean(), 3)
print(f"Mean sepal_length of versicolor observations: {mean_sepal_length}")

# how many species are present in this dataframe
species_count = iris['species'].nunique()
print(f"Number of species in the dataframe: {species_count}")

# count how many virginica
virginica_count = iris[iris['species'] == 'virginica'].shape[0]
print(f"Number of virginica observations: {virginica_count}")

```

```python
# 23

# What is the pearson correlation of sepal_length and sepal_width
sepal_length_sepal_width_corr = iris['sepal_length'].corr(iris['sepal_width'])
print(f"Pearson correlation between sepal_length and sepal_width: {sepal_length_sepal_width_corr}")

# What is the pearson correlation of petal_length and sepal_length
petal_length_sepal_length_corr = iris['petal_length'].corr(iris['sepal_length'])
print(f"Pearson correlation between petal_length and sepal_length: {petal_length_sepal_length_corr}")

# What is the pearson correlation of sepal_length petal_width
sepal_length_petal_width_corr = iris['sepal_length'].corr(iris['petal_width'])
print(f"Pearson correlation between sepal_length and petal_width: {sepal_length_petal_width_corr}")

# What is the pearson correlation of petal_length and petal_width
petal_length_petal_width_corr = iris['petal_length'].corr(iris['petal_width'])
print(f"Pearson correlation between petal_length and petal_width: {petal_length_petal_width_corr}")

# identify the highest out of all 4
highest_corr = max(sepal_length_sepal_width_corr, petal_length_sepal_length_corr, sepal_length_petal_width_corr, petal_length_petal_width_corr)
print(f"The highest correlation is: {highest_corr}")

```

```python
# 24

# what is the mean value of petal_length of setosa rounded to two decimals
setosa_petal_length_mean = round(iris[iris['species'] == 'setosa']['petal_length'].mean(), 2)
print(f"Mean petal_length of setosa: {setosa_petal_length_mean}")

# What has the largest average sepal_width
species_sepal_width_mean = iris.groupby('species')['sepal_width'].mean()
max_species_sepal_width_mean = species_sepal_width_mean.idxmax()
print(f"Species with the largest average sepal_width: {max_species_sepal_width_mean}")


# what is the standard deviation of sepal_length of setosa rounded to two decimals
setosa_sepal_length_std = round(iris[(iris['species'] == 'setosa')]['sepal_length'].std(), 2)
print(f"Standard deviation of sepal_length of setosa: {setosa_sepal_length_std}")

# what has the largest sepal_length
species_sepal_length_max = iris.groupby('species')['sepal_length'].max()
max_species_sepal_length = species_sepal_length_max.idxmax()
print(f"Species with the largest sepal_length: {max_species_sepal_length}")
```

```python
# 25

import seaborn as sns
titanic = sns.load_dataset('titanic')

# how many passengers in pclass of 2 has survived equal to 1 in the titanic dataset?
pclass_2_survived_count = titanic[(titanic['pclass'] == 2) & (titanic['survived'] == 1)].shape[0]
print(f"Number of passengers in pclass 2 who survived: {pclass_2_survived_count}")

```

```python
# 26
# What is the age range in pclass of 1 where survived = 0?

# Filter for 1st class passengers who did not survive
first_class_not_survived = titanic[(titanic['pclass'] == 1) & (titanic['survived'] == 0)]

# Get the minimum and maximum age
min_age = first_class_not_survived['age'].min()
max_age = first_class_not_survived['age'].max()

# Print the age range
print(f"Age range of 1st class passengers who did not survive: {min_age}-{max_age}")
```

```python
# 27
# Filter for female passengers who did survive
female_survived = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)]

# percentage of female passengers who survived, rounded to the nearest whole percent
female_survived_percentage = round(female_survived.shape[0] / titanic[titanic['sex'] == 'female'].shape[0] * 100)
print(f"Percentage of female passengers who survived: {female_survived_percentage}%")

```

```python
# 28
# filter through titanic dataset and count how many missing values there are?
missing_values_count = titanic.isnull().sum()
print(missing_values_count)

# identify which variable goes with the number of missing values
missing_values_count = missing_values_count.idxmax()
print(missing_values_count)

# display only the variables that have missing values in the titanic dataset?
variables_with_missing_values = titanic.columns[titanic.isnull().any()]
print(variables_with_missing_values)

```
